## Results of the __Dense NN__ tests.
__NOTE:__ A few modifications are applied to the results datafile for __Dense NN__ as they were recoreded in a slightly differenr fashion.

```python
# the recording of scores for dense-NN was done a little different to the other classifiers.
# previous tests had caused the vram to be fully used resulting in the testing process stopping
# without completing. a method of forcing keras to free the vram after every test by starting
# each test in its own process was used. but, this then made the comparing of results and only
# saving the best scores more difficult. so, all scores were recorded and the identifying of 
# the best scores is performed here.

# read the file containing all results
df_dense_nn_unformatted = pd.read_csv("".join([data_path, "Dense-NN_for_Tiles_Classifier.csv"]), header = None)

# apply some basic formatting to bring it inline with the other score files
df_dense_nn_unformatted.columns = (["classifier", "group", "params", "score", "target"])
df_dense_nn_unformatted["group"] = "TILES"
df_dense_nn_unformatted["target"] = df_dense_nn_unformatted["target"].apply(lambda x: groups[x])

# find the best scores for each taget
df_dense_nn_artist = df_dense_nn_unformatted[df_dense_nn_unformatted["target"] == "ARTIST"].sort_values("score", ascending = False).head(1)
df_dense_nn_genre  = df_dense_nn_unformatted[df_dense_nn_unformatted["target"] == "GENRE"].sort_values("score", ascending = False).head(1)
df_dense_nn_style  = df_dense_nn_unformatted[df_dense_nn_unformatted["target"] == "STYLE"].sort_values("score", ascending = False).head(1)

# combine the best scores into a single dataframe
df_dense_nn = pd.DataFrame()
df_dense_nn = df_dense_nn.append(df_dense_nn_artist)
df_dense_nn = df_dense_nn.append(df_dense_nn_genre)
df_dense_nn = df_dense_nn.append(df_dense_nn_style)
df_dense_nn
```

## Combine all of the Lists and generate 3 ordered lists for _Artist_, _Genre_ & _Style_.
__NOTE:__ These lists shall help us decide on the best classifiers to use in the final predictor. 

```python
df_results = pd.DataFrame()
df_results = df_results.append(df_knn_images)
df_results = df_results.append(df_knn_palettess)
df_results = df_results.append(df_knn_large_palettess)
df_results = df_results.append(df_logistic_regression)
df_results = df_results.append(df_random_forest)
df_results = df_results.append(df_svc)
df_results = df_results.append(df_dense_nn)

df_results_artist = df_results[df_results["target"] == "ARTIST"].sort_values("score", ascending = False)
df_results_genre  = df_results[df_results["target"] == "GENRE" ].sort_values("score", ascending = False)
df_results_style  = df_results[df_results["target"] == "STYLE" ].sort_values("score", ascending = False)
```

## Put the Information into Plots to Make it Easier to Read.
Prepare the data. 

```python
df_results_artist.reset_index(inplace = True)
df_results_genre.reset_index(inplace = True)
df_results_style.reset_index(inplace = True)

df_results_artist["classifier_and_group"] = df_results_artist.apply(lambda x: "".join([ x["classifier"], "- ", x["group"]]), axis = 1)
df_results_genre["classifier_and_group" ] = df_results_genre.apply(lambda  x: "".join([ x["classifier"], "- ", x["group"]]), axis = 1)
df_results_style["classifier_and_group" ] = df_results_style.apply(lambda  x: "".join([ x["classifier"], "- ", x["group"]]), axis = 1)

df_results_artist["best_score"] = df_results_artist[df_results_artist["score"] == max(df_results_artist["score" ])]["score"]
df_results_genre["best_score" ] = df_results_genre[df_results_genre["score"  ] == max(df_results_genre[ "score" ])]["score"]
df_results_style["best_score" ] = df_results_style[df_results_style["score"  ] == max(df_results_style[ "score" ])]["score"]

df_results_artist["best_score"].fillna(0, inplace = True)
df_results_genre["best_score" ].fillna(0, inplace = True)
df_results_style["best_score" ].fillna(0, inplace = True)

params_artist = df_results_artist[df_results_artist["score"] == max(df_results_artist["score"])]["params"].values
params_genre  = df_results_genre[df_results_genre["score"  ] == max(df_results_genre["score" ])]["params"].values
params_style  = df_results_style[df_results_style["score"  ] == max(df_results_style["score" ])]["params"].values
```

The function <code>params_to_formatted_string()</code> generates a formatted string containing the best Hyperparameters that are associated with the bast classifiers.

```python
def params_to_formatted_string(params):
    
    # switch single quotes to doble quote so that the params string can be converted to a dictionary
    params_dict = json.loads(params[0].replace("'", '"'))
    
    # begin the result string with some title text
    result = "HYPERPARAMETERS\n\n"

    # get all of the items from the dictionary
    keys_values = params_dict.items()

    # loop through all items extracting the dictionary keys and values
    for key, value in keys_values:
        # append the hyperparameter (key), the value and a newline character to the result string
        result = "".join([result, key, " : ", str(value), "\n"])

    # return the list of hyperparameters as a formatted string
    return result
```

Generate the three plots _side-by-side_. In each case indicating which classifier is the best for its target group and the best hyperparameters to use.

```python
# define the figure that we shall plot to
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_size_inches(32, 9)

# create a simple bar plot
ax1.bar(df_results_artist["classifier_and_group"], df_results_artist["score"], alpha = 0.2, color = "salmon")
ax1.bar(df_results_artist["classifier_and_group"], df_results_artist["best_score"], alpha = 0.2, color = "red")
ax1.set_xticklabels(df_results_artist["classifier_and_group"], rotation = 90,    fontsize = 15)
ax1.set_title("The Best Classifier & Group to identify the Artist")
ax1.set_ylabel("Score")
plt.figtext(0.28, 0.72, params_to_formatted_string(params_artist), fontsize=14)

# create a simple bar plot
ax2.bar(df_results_genre["classifier_and_group"], df_results_genre["score"], alpha = 0.2, color = "palegreen")
ax2.bar(df_results_genre["classifier_and_group"], df_results_genre["best_score"], alpha = 0.2, color = "green")
ax2.set_xticklabels(df_results_genre["classifier_and_group"], rotation = 90,    fontsize = 15)
ax2.set_title("The Best Classifier & Group to identify the Genre")
plt.figtext(0.555, 0.64, params_to_formatted_string(params_genre), fontsize=14)

# create a simple bar plot
ax3.bar(df_results_style["classifier_and_group"], df_results_style["score"], alpha = 0.2, color = "lightsteelblue")
ax3.bar(df_results_style["classifier_and_group"], df_results_style["best_score"], alpha = 0.2, color = "steelblue")
ax3.set_xticklabels(df_results_style["classifier_and_group"], rotation = 90,    fontsize = 15)
ax3.set_title("The Best Classifier & Group to identify the Style")
plt.figtext(0.83, 0.64, params_to_formatted_string(params_style), fontsize=14)

# view the figure with its three subplots
plt.show() 
```

__COMMENT:__ So, the best classifier for predicting __ARTISTS__ is the __SVC__ using the __IMAGE__ dataset. For the Genre & Style the __TILES__ dataset along with the __Dense NN__ classifier is the winner in both catagories.
